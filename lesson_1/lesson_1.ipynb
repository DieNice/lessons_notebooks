# Циклы
Циклы for предназначены для обхода коллекции (например, списка или кортежа) или итератора. Стандартный синтаксис выглядит так:

````python
collection = [1,2,3,4,5]

for element in collection:
    print(element)
````

Ключевое слово **continue** позволяет сразу перейти к следующей итерации цикла, не доходя до конца блока. Рассмотрим следующий код, который суммирует  
целые числа из списка, пропуская значения None:  

````python
sequence = [1, 2, None, 4, None, 5]
total = 0
for value in sequence:
	if value is None:
		continue
	total += value
total
````

Из цикла for можно выйти досрочно с помощью ключевого слова **break**. Следующий код суммирует элементы списка, пока не встретится число 5:  

````python
sequence = [1, 2, 0, 4, 6, 5, 2, 1]
total = 0
for value in sequence:
	if value == 5:
		break
	total = total + value
total
````

Цикл while состоит из условия и блока кода, который выполняется до тех пор,  пока условие не окажется равным False или не произойдет выход из цикла в результате предложения break:  

````python
х = 256  
total = 0  
while х > 0:  
	if total > 500:  
		break  
	total += х  
	х = х // 2
total
````

При написании программы неизбежно столкновение с ошибками программы.

````python
my_lst = [1,2,3,"hello",5,10,-1]
my_lst.sort()
````

Все данные в языке Python представлены объектами и отношениями между объектами. В предыдущем занятии мы рассмотрели основные типы данных Python и то как их применять.

При вызове класса как функции неявно вызывается конструктор этого класса. Специальная "скрытая" функция, которая выполняет некоторую логику перед созданием объекта.
Такой механизм есть и у встроенных типов Python. Например - список. Можно вызвать как `[]` так и как `list()`.

````python
class Counter:
	pass #pass-это заглушка или пустая реализация

x = Counter()
# Наградили объект новым атрибутом
x.count = 0
x.count += 1
````

Чтобы заранее предопределить некоторое состояние или поведение объекта конкретного класса, а не писать один и тот же повторяющийся код, можно переопределить конструктор класса. Т.е. определить конструктор класса.

````python
class Counter:
	def __init__(self):
		self.count = 0

x = Counter()
print(x.count)
x.count += 1
````

**Self** Это ссылка объекта класса самого на себя. Она необходима чтобы функцию связать с классом.

````python
# Создание класса с определенной конфигурацией

class Counter:
	def __init__(self, start = 0):
		self.count = start
x1 = Counter(10)
x2 = Counter()
print(x2.count)
print(x1.count)
````

````python
# Методы класса

class Counter:
	def  __init__(self):
		self.count = 0

	def inc(self):
		self.count += 1
		
	def reset(self):
		self.count = 0

x = Counter()
x.inc()
print(x.count)
````

````python
Counter.inc(x)
print(x.count)
````

````python
x.reset()
print(x.count)
````
